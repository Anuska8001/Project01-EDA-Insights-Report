STEP 2 : CLEAN THE DATA

```python
df['date_added'].dtype
```

```python
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['type'].value_counts()
df['country'].value_counts().head(10)
df['rating'].value_counts()
df['release_year'].value_counts().head(10)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['date_added', 'duration'], inplace=True)
```

```python
df.isnull().sum()
```

Data Cleaning Summary:
1. Converted date_added to datetime format.
2. Filled missing values in country, director, cast and rating columns.
3. Removed rows with missing values in date_added and duration.
4. Checked and removed duplicate records.

STEP 4 : VISUALIZATIONS

Chart 1
Movies vs TV Shows

```python
plt.figure(figsize=(6,5))
sns.countplot(x='type', data=df)
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

Chart 2 Top 10 Countries

```python
plt.figure(figsize=(12,6))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

Chart 3 Content Released Over Years

```python
plt.figure(figsize=(12,6))
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Content Release Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

Chart 4 Ratings Distribution

```python
plt.figure(figsize=(12,6))
sns.countplot(y='rating', data=df,
              order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.show()
```

Chart 5 Top Directors

```python
top_directors = df[df['director'] != 'Unknown']['director'] \
                  .value_counts().head(10)

plt.figure(figsize=(12,6))
top_directors.plot(kind='barh')
plt.title('Top 10 Directors on Netflix')
plt.show()
```

Chart 6 Content Added Per Year

```python
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(12,6))
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title('Content Added to Netflix Each Year')
plt.xlabel('Year')
plt.ylabel('Titles Added')
plt.show()
```

STEP 5 : INSIGHTS REPORTS

Insight 1

Netflix has significantly more Movies (6,128) than TV Shows (2,666), indicating that movies are the primary type of content on the platform.

Insight 2

The United States dominates Netflix's content library with 2,809 titles, followed by India with 972 titles.

Insight 3

The year 2018 had the highest number of releases on Netflix with 1,146 titles, showing a major expansion of content during this period.

Insight 4

TV-MA is the most common rating category with 3,209 titles, suggesting that a large portion of Netflix content is targeted toward mature audiences.

Insight 5

Among known directors, Rajiv Chilaka has the highest number of titles (19), showing that only a few directors contribute multiple titles to Netflix.

I was surprised to see that movies outnumber TV shows by such a large margin on Netflix. Another interesting finding was the dominance of the United States in content production. The large amount of mature-rated content and the huge increase in releases around 2018 were also unexpected.

CONCLUSION

This exploratory data analysis revealed important trends in Netflix's content library. The platform is heavily dominated by movies and content from the United States. Most content is targeted toward mature audiences, and Netflix experienced significant growth in content releases between 2016 and 2019.
